```python
#< ----------------------   6 -List all the cities that hosted any olympic games more than once ------------------------------>



counts = df['City'].value_counts()
df[df['City'].isin(counts.index[counts > 1])].drop_duplicates(subset='City')
```

```python
#< ----------------------   7 -Plot histograms of Age, Height and Weight of athletes ------------------------------>





sns.set()   # to display in a nice format
plt.hist(df.Age.dropna(), alpha=0.9, color='Red', bins=20)
plt.xlabel("Athlets Age")
```

```python
sns.set()   # to display in a nice format
plt.hist(df.Height.dropna(), alpha=0.9, color='Green', bins=10)
plt.xlabel("Athlets Heights")
```

```python
sns.set()   # to display in a nice format
plt.hist(df.Weight.dropna(), alpha=0.9, color='Blue', bins=30)#dropna() functions removes the NaN values making possible to plot
plt.xlabel("Athlets Weight")
```
